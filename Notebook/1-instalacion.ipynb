<h1> Java y Maven en VS Code </h1>


<h2> Requisitos: </h2>

<h3> Instalar Java JDK </h3>

1. Descargar desde [Oracle JDK](https://www.oracle.com/java/technologies/javase-jdk.html) e instalar
2. Verificar la instalación en PowerShell o CMD con:
java --version

<h3>Debería mostrar algo similar a esto:</h3>

<pre>
java version "24.0.1" 2025-04-15
Java(TM) SE Runtime Environment (build 24.0.1+9-30)
Java HotSpot(TM) 64-Bit Server VM (build 24.0.1+9-30, mixed mode, sharing)
</pre>

Indicando que la instalación fue exitosa.

<h3>Instalar Apache Maven (programa, no extensión)</h3>

1. Descargar la versión Binary zip de Maven desde [Apache Maven Project](https://maven.apache.org/download.cgi)
2. Descomprimir dicho archivo en algún lugar a la mano como el escritorio.
3. Pasar la carpeta descomprimida a:

    <pre>C:\Program Files</pre>

La ruta en cuestión debería quedar similar (salvando los versionados) a esta:
<pre>   C:\Program Files\apache-maven-3.9.12</pre>



<h3>Configurar variables de entorno</h3>

1. Abrir Variables de entorno del sistema desde el navegador de windows.
2. Seleccionar "Editar las variables del entorno del sistema (no las de usuario).
3. Una vez dentro de las propiedades del sistema, elegir variables de entorno:



4. Una vez en la nueva ventana, seleccionar nueva:



5. Dentro de la ventana hay que agregar (nuevamente, la versión de maven puede ser distinta), MAVEN_HOME y C:\Program Files\apache-maven-3.9.12



6. Seleccionar la variable path y editarla



7. En la nueva ventana, seleccionar "Nuevo" y escribir:
<pre>
    %MAVEN_HOME%\bin
</pre>

<h3>Aconsejo cerrar todas las instancias previamente abiertas de powershell, cmd y vs code para que el sistema detecte los cambios realizados</h3>  

8. Luego de este proceso, seleccionar aceptar en las ventanas abiertas y probar los siguientes comandos en powershell:
mvn -v
$env:MAVEN_HOME

<h3>Debería mostrar algo similar a esto:</h3>

mvn-v
<pre>
Apache Maven 3.9.12 (848fbb4bf2d427b72bdb2471c22fced7ebd9a7a1)
Maven home: C:\Program Files\apache-maven-3.9.12
Java version: 24.0.1, vendor: Oracle Corporation, runtime: C:\Program Files\Java\jdk-24
Default locale: es_UY, platform encoding: UTF-8
OS name: "windows 11", version: "10.0", arch: "amd64", family: "windows"
</pre>

$env:MAVEN_HOME
<pre>
C:\Program Files\apache-maven-3.9.12
</pre>

<h3>Finalmente, instalar "maven for java" de oracle corporation y "extension pack for java" de microsoft en vscode. Reiniciar dicho programa y queda listo el proceso para crear proyectos en java.</h3>
