<h3>Primer proyecto java con maven</h3>

1. primero, con ctrl+shift+p para abrir la consola, escribir "Maven: new"

2. Luego la consola nos va a preguntar que "arquetipo se desea". Esto quiere decir que "template" o "paleta" se busca usar. Con esto en cuenta, vamos a usar "maven-archetype-quickstart" para aprender a manejar dependencias y no solo java.

3. Selecciona la versión 1.4 (luego la actualizaremos en el pom).

4. Ahora algún identificador del dueño del proyecto, por ejemplo com.ejercicios.seba.

5. Después algún identificador del proyecto en sí, por ejemplo ejercicios-java.

6. Luego se va a abrir una ventana que te va a pedir donde guardar el proyecto, selecciona algún lugar de fácil acceso.

cuando la consola pregunte:
<pre> Define value for property 'version' 1.0-SNAPSHOT:</pre>

esta pidiendo el "versionado", 1.0 significa la primera versión y snapshot significa que está en desarrollo e inestable. En este caso se puede simplemente dejar así.

si se quieren saber las convenciones están aca: [Apache Maven](https://maven.apache.org/guides/mini/guide-naming-conventions.html)

7. Luego, va a aparecer un mensaje similar a este:
<pre>
Confirm properties configuration:
groupId: com.ejercicios.seba
artifactId: ejercicios-java
version: 1.0-SNAPSHOT
package: com.ejercicios.seba
 Y: :
</pre>

simplemente presiona enter para confirmar.

<h3>
Ahora hay que configurar el .pom
</h3>

Donde dice "compiler.source" y "compiler.target" cambiarlos a 1.8. (o la versión más estable de java del momento)


<h3> Ahora solo falta correr el programa </h3>

1. Desde el pom, abre la consola y escribe:
<pre> mvn package </pre>

esto va a formar el package que luego vamos a compilar.


2. luego que se forme el package, escribe:
java -cp target/classes com.ejercicios.seba.App

3. Ahora vamos a poder ver nuestro gran código.
