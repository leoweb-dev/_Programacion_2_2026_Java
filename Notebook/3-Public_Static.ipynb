<h3>¿Qué es public static void main(String[] args) en Java? </h3>

En Java, el método:

<h4>public static void main(String[] args)</h4>

- Es el punto de entrada del programa, es decir: donde se ejecuta.

<h4>public</h4>

- Significa que el método es público, es decir que se puede acceder al mismo fuera de la clase que lo creó. Si no se hace público, Java no lo va a poder ejecutar.

<h4>static</h4>

- Significa que pertenece a la clase (que lo creó) y no a un objeto (de ninguna clase). No se necesita crear ningún objeto para ejecutarlo.

    - Recordemos que un objeto es una instancia de una clase, es decir un "ejemplar" creado a partir de los parámetros definidos en la clase. Para poner el ejemplo clásico: la clase es Perro y el objeto es miPerro, un perro concreto con sus propias características.

<h4>void</h4>

- Significa que el método en concreto no va a retornar nada.

    - Una aclaración importante: void indica que el método main no puede retornar ningún valor al llamarlo desde otro método. No se puede hacer return dentro de main, porque eso entra en conflicto con el "void". Solo se puede hacer return; vacío, el cual termina el método sin devolver nada.

<h4>main</h4>

- Es el nombre que Java reconoce como punto de inicio (similar a index en las páginas web). Si se cambia por otro nombre, el programa no inicia.

<h4>(String[] args)</h4>

- Significa que es un arreglo de String. Sirve para recibir argumentos desde la línea de comandos (la consola).

    - Esto no quiere decir que solo se pueda trabajar con String, sino que es como se van a recibir los argumentos para que Java corra el programa.

<h1>Breve recordatorio:</h1>

<h2>Objeto</h2>
<h3>Es una entidad concreta en memoria que agrupa datos (atributos) y comportamientos (métodos) definidos por una clase. Representa "algo" que existe durante la ejecución del programa.</h3>
<h2>Instancia</h2>
<h3>Es la creación particular de una clase en un momento dado. Cuando se “instancia” una clase, se está generando un objeto en memoria con las propiedades y métodos definidos por esa clase.</h3>
